```python
import sys
reload(sys)
sys.setdefaultencoding("utf-8")
import nltk
from nltk import tokenize
from gensim.models import word2vec
from sklearn.decomposition import PCA
from bokeh.plotting import figure, output_file, show, ColumnDataSource, vplot
from bokeh.io import output_notebook
output_notebook()
```

```python
corpus = tokenize.sent_tokenize(''.join(open('bibliaUTF8.txt').readlines()))
```

```python
corpus = map(nltk.word_tokenize,corpus)
```

```python
frequency = {}
for s in corpus:
    for w in s:
        if w not in frequency:
            frequency[w] = 0
        else:
            frequency[w] +=1
```

```python
model = word2vec.Word2Vec(corpus,size=120,window=5,min_count=50)
```

```python
model.most_similar(positive=['Dios'])
```

```python
model.most_similar(positive=['amor'])
```

```python
model.most_similar(positive=['guerra'])
```

```python
words = []
vectors = []
for s in corpus:
    for w in s:
        if w in model and w not in words and frequency[w] > 50:
            words.append(w)
            vectors.append(model[w])
```

```python
len(words)
```

```python
pca = PCA(n_components=2)
vectors_pca = pca.fit_transform(vectors)
```

```python
x = []
y = []
label = []
for idx,v in enumerate(words):
    label.append(v)
    x.append(vectors_pca[idx][0])
    y.append(vectors_pca[idx][1])
```

```python
source = ColumnDataSource(data=dict(x=x,y=y, label=label))
#Nodes Plotting
o = figure(plot_height=400,plot_width=800)
o.text('x', 'y', label, source=source)
o.circle('x', 'y', size=10, source=source,alpha=0.5 )
show(o)
```
